```python
import matplotlib.pyplot as plt
import pickle
with open('./../results/brightkite_10k_nodes/link_SAGE_add_brightkite_layer3_approximate-1.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
import numpy as np
data = np.array(data)
timesteps = data[:,1]
scores = data[:,0][0]
scores = np.array(scores)
auc_scores = scores[:,0]
prec_scores = scores[:,1]
f1_scores = scores[:,2]
rec_scores = scores[:,3]
```

```python
fig = plt.figure()
plt.title('Metrics vs Training time')
plt.suptitle("GraphSAGE on Brightkite dataset")
plt.plot(timesteps, auc_scores, label='AUC-ROC' )
plt.plot(timesteps, prec_scores, label='Precision' )
plt.plot(timesteps, f1_scores,label='F1-score' )
plt.plot(timesteps, rec_scores,label='Recall' )
plt.xlabel('Time (in seconds)')
plt.ylabel('metric')
plt.legend(loc="upper right")
plt.show()
```
